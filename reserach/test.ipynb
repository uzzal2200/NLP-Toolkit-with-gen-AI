```python
import google.generativeai as genai
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
# Access the API key from environment
# Check GOOGLE_API_KEY first, then fallback to GEMINI_API_KEY
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY") or os.getenv("GEMINI_API_KEY")

print("✓ API key loaded successfully")
```

```python
user_text1 = """
Give me the sentiment of this sentence:
"I hate this movie"
"""

user_text2 = """
Give me hindi transilation of this sentence:
"I hate this movie"
"""


user_text3 = """
Detect the language of this sentence:
"I hate this movie"
"""
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel("gemini-2.5-flash")
response = model.generate_content(user_text3)
results = response.text
print(results)
```
